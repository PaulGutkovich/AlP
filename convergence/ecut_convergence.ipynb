```python
import matplotlib.pyplot as plt
```

```python
def visualize_energies(filename, save=None):
    cutoffs = []
    energies = []
    with open(filename, "r") as f:
        lines = f.read().split("\n")
        n_lines = len(lines)
        for i in range(n_lines//2):
            cutoff = float(lines[2*i].split(" ")[-1])
            energy = float(lines[2*i+1].split(" ")[-2])
            cutoffs.append(cutoff)
            energies.append(energy)
    plt.plot(cutoffs, energies, "o-", markersize=5, label='Etot vs ecutwfc')
    plt.xlabel('ecutwfc (Ry)')
    plt.ylabel('Etot (Ry)')
    plt.legend(frameon=False)
    if save is not None:
        plt.savefig(save)
```

```python
visualize_energies("ecut_energies.txt", "ecut_energies.png")
```
